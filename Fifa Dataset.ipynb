```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming `df` is your DataFrame with an 'age' column
# Create age ranges with 5-year gaps
age_ranges = range(df['Age'].min(), df['Age'].max() + 6, 5)
labels = [f"{i}-{i+4}" for i in age_ranges[:-1]]
df['Age_range'] = pd.cut(df['Age'], bins=age_ranges, labels=labels)

# Create histogram with age ranges
plt.hist(df['Age'], bins=age_ranges, edgecolor='black', alpha=0.75)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution with 5-Year Gap')

# Add total number on top of bars
for i in range(len(age_ranges) - 1):
    count = df['Age_range'].value_counts().sort_index()[labels[i]]
    plt.text(age_ranges[i] + 2.5, count + 2, str(count), ha='center')

plt.show()
```

```python
import pandas as pd

# Load DataFrame
  # Replace with your actual DataFrame

# Define columns to convert to numeric
columns_to_convert = ['Value', 'Wage', 'Height', 'Weight']  # Replace with column names to be converted

# Loop through columns and convert to numeric
for col in columns_to_convert:
    # Remove non-numeric characters from column values
    df[col] = df[col].str.replace('[^\d.]', '', regex=True)
    # Convert column to float
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Print updated data types of columns
print(df[columns_to_convert].dtypes)
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Stamina','SprintSpeed','HeadingAccuracy','LongShots']], df['Overall'], test_size=0.25, random_state=1)

model = LinearRegression()
model.fit(x_train, y_train)

score = model.score(x_test, y_test)
print('model score', score)

# Make predictions on new data
new_data = pd.DataFrame({'Stamina': [1, 2], 'SprintSpeed': [3, 4], 'HeadingAccuracy': [5, 6], 'LongShots': [7, 8]})
predictions = model.predict(new_data)

print('Predictions:', predictions)

```

```python

# Define the dependent variable (y) and independent variables (x1, x2, x3)
y = df['Overall']
x1 = df['GKDiving']
x2 = df['GKHandling']
x3 = df['GKKicking']
x4 = df['GKPositioning']
x5 = df['GKReflexes']
x6 = df['Penalties']

# Combine the independent variables into a DataFrame
X = pd.DataFrame({'GKDiving': x1, 'GKHandling': x2, 'GKKicking': x3, 'GKPositioning': x4, 'GKReflexes': x5, 'Penalties': x6})

# Drop any rows with missing values
X = X.dropna()

# Add a constant term to the independent variables to represent the intercept
X = sm.add_constant(X)

# Create the OLS regression model
model = sm.OLS(y, X)

# Fit the model
results = model.fit()

# Print the summary statistics
print(results.summary())

# Make predictions on new data
new_data = pd.DataFrame({'GKDiving': [1, 2], 'GKHandling': [3, 4], 'GKKicking': [5, 6], 'GKPositioning': [7, 8], 'GKReflexes': [9, 10], 'Penalties': [11, 12]})
new_data = sm.add_constant(new_data)
predictions = results.predict(new_data)

print('Predictions:', predictions)
```
